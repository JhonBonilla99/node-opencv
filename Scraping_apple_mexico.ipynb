```python
#importamos Bibliotecas
from bs4 import BeautifulSoup
import requests
```

```python
import bs4
print("version de Beatiful soup", bs4.__version__)
print("version de requests", requests.__version__)
```

```python
#variables del programa
url_apple_mexico = "https://jobs.apple.com/en-us/search?sort=relevance&location=mexico-MEXC"
```

```python
#hacemos solicitud a la pagina
pedido_al_servidor = requests.get(url_apple_mexico)

#obtenemos el html con .text
html_apple = pedido_al_servidor.text
```

```python
#parseamos el HTML

soup = BeautifulSoup(html_apple, "html.parser")
```

```python

#obtenemos los elementos <a> que contienen los links a los detalles de los empleos
ids = soup.find_all('a', class_ = "table--advanced-search__title")
```

```python
for a in ids:
  print(a)
  print(" ")
```

```python
#obtener los links que estan en los href de cada empleo, tambien los guardamos en una lista
hrefs = [a.get('href') for a in ids]
```

```python
for href in hrefs:
    print(href)
```

```python
#armar los links de cada empleo con la estructura https://jobs.apple.com[href]
links_empleos = ["https://jobs.apple.com"+ href for href in hrefs]
```

```python
#imprimimos los links
for link in links_empleos:
  print(link)
```

```python
#Dataframe
from typing import List, Optional
from pydantic import BaseModel, Extra
import json
```

```python


class Skill(BaseModel):
    name: str
    level: str
    experience: int

class Aptitude(BaseModel):
    name: str

class Tool(BaseModel):
    name: str
    level: str
    experience: int

class Language(BaseModel):
    name: str
    level: str

class Benefit(BaseModel):
    name: str

class JobPost(BaseModel):
    name: Optional[str] = None
    externalLink: Optional[str] = None
    description: Optional[str] = None
    createdAt: Optional[str] = None
    availableSlots: Optional[int] = None
    skills: Optional[List[Skill]] = None
    aptitudes: Optional[List[Aptitude]] = None
    tools: Optional[List[Tool]] = None
    languages: Optional[List[Language]] = None
    benefits: Optional[List[Benefit]] = None
    scholarity: Optional[str] = None
    workhours: Optional[int] = None
    locationConditions: Optional[str] = None
    nationalRemote: Optional[str] = None
    minSalary: Optional[int] = None
    maxSalary: Optional[int] = None
    minAge: Optional[int] = None
    maxAge: Optional[str] = None
    sex: Optional[str] = None
    yearsOfExperience: Optional[int] = None
    status: Optional[str] = None
    country: Optional[str] = None
    updatedAt: Optional[str] = None
    driversLicense: Optional[bool] = None
    degree: Optional[bool] = None
    validPassport: Optional[bool] = None
    validVisa: Optional[bool] = None
    nationalRelocation: Optional[bool] = None
    internationalRelocation: Optional[bool] = None
    availabilityToTravel: Optional[bool] = None
    seniority: Optional[str] = None
    showSalaryRange: Optional[bool] = None
    state: Optional[str] = None
    city: Optional[str] = None
    postalCode: Optional[str] = None
    slug: Optional[str] = None
    latitude: Optional[float] = None
    longitude: Optional[float] = None
    companyName: Optional[str] = None
    companyImg: Optional[str] = None
    class Config:
        extra = Extra.ignore
```

```python
#orden de obtencion de datos para el dataframe
#name
#description
#escolarity
#country
#state
#city
```

```python
job_posts = []
print(job_posts)
```

```python
for linkempleo in links_empleos:
  print(linkempleo)
  #prueba para obtener el html de un empleo
  pedido_empleo = requests.get(linkempleo)
  #obtenemos el html
  html_empleo = pedido_empleo.text
  #parseamos el html
  soup_job = BeautifulSoup(html_empleo, "html.parser")
  #obtenemos name job
  name_job = soup_job.find('h1', class_ = "jd__header--title").text
  print(name_job)

  #buscamos titulos con el h2 para cada seccion del empleo
  detalles_empleo = soup_job.find('div', class_ = "accordion accordion--job-details")
  #buscamos las secciones por secciones
  sections = detalles_empleo.find_all('section')
  #obtenemos description job
  description_job = sections[1].find('div', id = "jd-description").text
  print(description_job)
  #Obtenemos escolarity_job
  if sections[2].find('span').text == "":
    education = None
  else:
    education = sections[2].find('span').text
    print(education)
  escolarity_job = education

  #Obtenemos country que siempre va a ser mexico
  country_job = "Mexico"

  #obtenemos city y state
  #ubicacion
  location = soup_empleo.find('div', id="job-location-name").text
  #city
  if location == "Mexico":
    city_job = "Various Locations within Mexico"
  else:
    city_job = location
  print(city_job)

  #state
  if city_job == "Various Locations within Mexico":
    state_job = "Various Locations within Mexico"
  else:
    state_job = location

  #pasar datos al dataframe
  job_post = JobPost(
    name= name_job,
    description=description_job,
    scholarity=escolarity_job,
    country=country_job,
    state= state_job,
    city= city_job
)

  #Guardar el trabajo en un dataframe general
  job_posts.append(job_post.json())
```

```python
print(job_posts)
```

```python



dict_list = [json.loads(item) for item in job_posts]

# Mostrar la lista de diccionarios
print(dict_list)
```

```python
json_string = json.dumps(dict_list)
```

```python
print(json_string)
```
